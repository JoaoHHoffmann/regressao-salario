```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/Salary.csv')
df.head()
```

## Regressão Linear
Realizando uma análise preditiva de uma salário baseado nos seus anos de experiência.
*   Variável dependente: Salário
*   Variável Independente: Anos de experiência



```python
# Separando os dados para as variaveis independentes e dependente
Xv = df.loc[:, 'YearsExperience'].values
y = df.loc[:, 'Salary'].values
```

```python
#Converter X, vetor n elementos, para uma matriz de n x 1
Xm = np.array([Xv])
X = Xm.transpose()
print(X)
```

```python
# Criando o objeto para a regressão com sklearn
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
# Realizando o ajuste do modelo
reg.fit(X,y)
```

```python
# Obtendo os valores ajustados
y_estimado = reg.predict(X)
```

```python
# Visualizando a reta e a base de dados
plt.scatter(X, y, color = 'blue')
plt.plot(X, y_estimado, color = 'red')
plt.title('Análise preditiva de salário baseada nos anos de experiencia')
plt.xlabel('Anos de experiência')
plt.ylabel('Salário')
plt.show()
```
